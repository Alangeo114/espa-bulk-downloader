```python
import mongoengine
from mongoengine import Document
from mongoengine import DateTimeField
from mongoengine import DictField
from mongoengine import EmailField
from mongoengine import EmbeddedDocument
from mongoengine import EmbeddedDocumentField
from mongoengine import IntField
from mongoengine import ListField
from mongoengine import ReferenceField
from mongoengine import StringField
from mongoengine import URLField

from datetime import datetime
########################################################################################################################
# Default product options
########################################################################################################################
default_cdr_ecv_options = {
    'include_sourcefile'      : False,    #delivers underlying raster product as part of order
    'include_source_metadata' : False,    #delivers underlying metadata
    'include_sr_toa'          : False,    #deliver LEDAPS based top of atmosphere
    'include_sr_thermal'      : False,    #deliver LEDAPS band 6
    'include_sr'              : False,    #deliver LEDAPS surface reflectance
    'include_sr_browse'       : False,    #generate and deliver surface reflectance browse image
    'include_sr_ndvi'         : False,    #deliver normalized difference vegetation
    'include_sr_ndmi'         : False,    #deliver normalized difference moisture
    'include_sr_nbr'          : False,    #deliver normalized burn ratio
    'include_sr_nbr2'         : False,    #deliver normalized burn ratio 2
    'include_sr_savi'         : False,    #deliver soil adjusted vegetation
    'include_sr_msavi'        : False,    #deliver modified soil adjusted vegetation
    'include_sr_evi'          : False,    #deliver enhanced vegetation
    'include_swe'             : False,    #deliver surface water extent
    'include_sca'             : False,    #deliver snow covered area
    'include_solr_index'      : False,    #deliver a solr search index record
    'include_cfmask'          : False     #deliver cfmask in a seperate file (normally delivered in sr)
}

########################################################################################################################
# Default product options
########################################################################################################################
default_lpvs_options = {}


########################################################################################################################
# Default projection options
########################################################################################################################
def get_default_projection_options():
    '''returns default options for reprojecting scenes'''
    options = {}
    #reprojection options
    options['reproject'] = False                          #reproject all rasters (True/False)
    options['target_projection'] = None                   #if 'reproject' which projection?
    options['central_meridian'] = None                    #
    options['false_easting'] = None                       #
    options['false_northing'] = None                      #
    options['origin_lat'] = None                          #
    options['std_parallel_1'] = None                      #
    options['std_parallel_2'] = None                      #
    options['datum'] = 'wgs84'                            #
    
    #utm only options
    options['utm_zone'] = None                            # 1 to 60
    options['utm_north_south'] = None                     # north or south
    
    return options


########################################################################################################################
# Default subset options
########################################################################################################################
def get_default_subset_options():
    '''return default options for subsetting and framing'''
    options = {}
    #image framing/subsetting options
    options['image_extents'] = False                      #modify image extents (subset or frame)
    options['minx'] = None                                #
    options['miny'] = None                                #
    options['maxx'] = None                                #
    options['maxy'] = None                                #    
    return options


########################################################################################################################
# Default resize options
########################################################################################################################
def get_default_resize_options():
    '''return default options for resizing'''
    options = {}
    #Pixel resizing options
    options['resize'] = False                             #resize output product pixel size (True/False)
    options['pixel_size'] = None                          #if resize, how big (valid range 30 to 1000 meters)
    options['pixel_size_units'] = None                    #meters or dd.     
    return options


########################################################################################################################
# Default resample options
########################################################################################################################
def get_default_resample_options():
    '''returns default options required for resampling'''
    options = {}
    #Must have this when reprojecting or resizing pixels
    options['resample_method'] = 'near'                     #for ops that need it, how would user like to resample?
    return options



########################################################################################################################
# Default options
########################################################################################################################
def get_default_options(system):
    if system == "cdr_ecv":
        opts = default_cdr_ecv_options
        opts.update(get_default_product_options())
        opts.update(get_default_projection_options())
        opts.update(get_default_subset_options())
        opts.update(get_default_resize_options())
        opts.update(get_default_resample_options())    
        return opts
    elif system == "lpvs":
        return default_lpvs_options
    else:
        return dict()
    
    
    

class EventLog(Document):
    timestamp = DateTimeField(default=datetime.now(), db_field="ts")
    #location = StringField(db_field="loc")
    #user = StringField(db_field="u")
    #scene = StringField(db_field="s")
    #order = StringField(db_field="o")
                        
    category = StringField(db_field="cat")
    entry = DictField(db_field="etr")

class User(Document):
    email = EmailField(unique=True, db_field="em")
    registration_id = IntField(db_field="regid")
    events = ListField(ReferenceField(EventLog, reverse_delete_rule=mongoengine.NULLIFY), db_field="evt")

class Distribution(EmbeddedDocument):
    server = StringField(db_field="srvr")
    disk_location = StringField(db_field="dl")
    url = URLField(db_field="u")
    
class Checksum(EmbeddedDocument):
    value = StringField(db_field="val")
    bytes = StringField(db_field="b")
    disk_location = StringField(db_field="dl")
    url = URLField(db_field="u")
    
class Scene(EmbeddedDocument):
    name = StringField(db_field="n")
    events = ListField(ReferenceField(EventLog), db_field="evt")
    distribution = EmbeddedDocumentField(Distribution, db_field="dist")
    checksum = EmbeddedDocumentField(Checksum, db_field="chk")

class Order(Document):
    name = StringField(unique=True, db_field="n")
    system = StringField(choices=('cdr_ecv', 'lpvs'), db_field="sys")
    source = StringField(choices=('bulk_ordering', 'api', 'ee'), db_field="src")
    status = StringField(choices=('ordered', 'processing', 'complete'), db_field="stat")
    stage = StringField(choices=('hadoop', 'composite'), db_field="stg")
    scenes = ListField(EmbeddedDocumentField(Scene), db_field="scn")
    options = DictField(default=get_default_options(system), db_field="opt")
    events = ListField(ReferenceField(EventLog, reverse_delete_rule=mongoengine.NULLIFY), db_field="evt")
    note = StringField(db_field="nt")

    
with mongoengine.connect("mytestdb") as client:
    l1 = EventLog(category="scene_start", entry={"location":"l8srlscp32", "l1t_filesize":1234456}).save()
    l2 = EventLog(category="ledaps_start", entry={"input_filesize":342234234, "free_disk_space":987987987}).save()
    l3 = EventLog(category="ledaps_end").save()
    l4 = EventLog(category="scene_end").save()
    
    scene = Scene(name="test_scene", events = [l1, l2, l3, l4])
    order = Order(name="test_order", scenes=[scene]).save()
    
```
